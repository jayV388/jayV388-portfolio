# COVID-19 Global Data Analysis


Exploratory data analysis of COVID-19 trends across 248 countries and 6 continents using the [Our World in Data](https://ourworldindata.org/covid-deaths) dataset.

---

## Dataset

- **File:** `covid_data.csv`
- **Rows:** 262,423 | **Columns:** 67
- **Date range:** January 2020 – March 2023
- **Source:** Our World in Data

---

## Setup

Install dependencies:

```bash
pip install pandas numpy matplotlib seaborn
```

Open the notebook:

```bash
jupyter notebook covid_analysis.ipynb
```

# COVID-19 Data Analysis & Visualization

## 📊 Project Overview
This project provides a comprehensive analysis of COVID-19 data across countries and continents. It includes data cleaning, exploratory data analysis (EDA), and multiple visualizations to understand pandemic trends, death rates, and geographic distribution.

## Key Findings

- **Total cases:** ~715M across 231 countries
- **Global death rate:** ~1%
- **Most affected:** United States (cases), Africa (highest death rate %)
- **Peak month:** January 2022
- **Strongest correlation:** Higher GDP / life expectancy → more cases per 100k (likely due to better reporting)
